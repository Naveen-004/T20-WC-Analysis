```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
import sys
print(sys.setrecursionlimit(5000))
```

```python
response = requests.get("https://www.espncricinfo.com/rankings/content/page/211271.html")
response.ok
```

```python
content = response.content
read = BeautifulSoup(content, "html.parser")
```

```python
read.div
```

```python
div = read.find_all('div', class_='ciPhotoContainer')
len(div)
```

```python
div
```

```python
formats = div[0].find_all('h3')
formats
```

```python
for i in formats:
    print(i.text)
```

```python
read.table
```

```python
all_tables = read.find_all("table" ,  class_="StoryengineTable")
len(all_tables)
```

```python
odi = all_tables[1]
```

```python
col = []
for i in odi.find_all("th"):
    col.append(i.text)

col
```
