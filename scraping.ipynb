```python
response = requests.get("https://stats.espncricinfo.com/ci/engine/records/team/match_results.html?id=2022%2F23;trophy=89;type=season")
soup = BeautifulSoup(response.content, 'html.parser')
urls = []
links = soup.find_all('a', class_='data-link', href=True)
for link in links:
    urls.append(link.get('href'))
urls
```

```python
filtered_urls = []

for url in urls:
  if "ci/engine/match" in url:
    filtered_urls.append(url)

print(len(filtered_urls))
filtered_urls
```

## Batting Table

```python
data = pd.DataFrame(columns=['BATTING', 'Unnamed: 1', 'R', 'B', 'M', '4s', '6s', 'SR', 'Unnamed: 8',
       'Unnamed: 9', 'match', 'team'])
for url in filtered_urls:
    res = requests.get("https://stats.espncricinfo.com"+ url)
    soup = BeautifulSoup(res.content, 'html.parser')
    div = soup.find_all('div', class_ = 'ds-flex ds-items-center ds-min-w-0 ds-mr-1')
    div = div[-2].text +' vs '+ div[-1].text
    span = soup.find_all('span', class_='ds-text-title-xs ds-font-bold ds-capitalize')
    tables = soup.find_all('table', class_='ci-scorecard-table')
    c = 0
    for table in tables:
        df = pd.read_html(str(table))
        df = pd.DataFrame(df[0])
        df['match'] = df.apply(lambda row: div, axis=1)
        df['team'] = df.apply(lambda row: span[c].text, axis=1)
        c = c+1
        data = pd.concat([data,df], axis=0, ignore_index=True)
data  
```

```python
data.to_csv('data/battingsummary.csv', index=False)
```

## Bowling Table

```python
data = pd.DataFrame()
for url in filtered_urls:
    res = requests.get("https://stats.espncricinfo.com"+ url)
    soup = BeautifulSoup(res.content, 'html.parser')
    tables =  soup.find_all('table', class_='ds-w-full ds-table ds-table-md ds-table-auto')
    div = soup.find_all('div', class_ = 'ds-flex ds-items-center ds-min-w-0 ds-mr-1')
    div = div[-2].text +' vs '+ div[-1].text
    span = soup.find_all('span', class_='ds-text-title-xs ds-font-bold ds-capitalize')
    c=1
    for table in tables:
        df = pd.read_html(str(table))
        df = pd.DataFrame(df[0])
        df['match'] = df.apply(lambda row: div, axis=1)
        df['team'] = df.apply(lambda row: span[c].text, axis=1)
        c = c-1
        data = pd.concat([data,df], axis=0, ignore_index=True)
data
```

```python
data.to_csv('data/bowlingsummary.csv', index=False)
```

## Players

```python
player_urls = []
for url in filtered_urls:
    res = requests.get("https://stats.espncricinfo.com"+ url)
    soup = BeautifulSoup(res.content, 'html.parser')
    links = soup.find_all('a', href=True)
    for link in links:
        if '/player/' in link.get('href'):
            player_urls.append(link.get('href'))
player_urls
```

```python
from itertools import chain # Flatten the list library
temp = pd.DataFrame(player_urls) # Convert the list to a dataframe
temp.drop_duplicates(inplace=True) # Remove duplicates
player_urls = temp.to_numpy().tolist() # Convert the dataframe to a list
player_urls = list(chain(*player_urls)) # Flatten the list
player_urls
```

```python
player_info = pd.DataFrame(columns=['name', 'team', 'batting_style', 'bowling_style', 'player_role'])
for url in player_urls:
    res = requests.get("https://espncricinfo.com"+ url)
    soup = BeautifulSoup(res.content, 'html.parser')
    name = soup.find_all('h1', class_='ds-text-title-l ds-font-bold')
    team = soup.find_all('span', class_='ds-text-comfortable-s')
    h5 = soup.find_all('h5')
    
    temp = {'name':name[0].get_text(strip=True) if name[0] else None, 'team': team[0].get_text(strip=True) if team[0] else None, 
    'batting_style': h5[3].get_text(strip=True) if h5[3] else None, 'bowling_style': h5[4].get_text(strip=True) if h5[4] else None, 
    'player_role': h5[5].get_text(strip=True) if h5[5] else None}
    print(temp)
    temp = pd.DataFrame(temp, index=[0])
    player_info = pd.concat([player_info, temp], ignore_index=True)
        
player_info
```

```python
player_info.to_csv('data/playerinfo.csv', index=False)
```
