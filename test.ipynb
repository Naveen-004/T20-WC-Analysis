```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
response = requests.get("https://stats.espncricinfo.com/ci/engine/records/team/match_results.html?id=2022%2F23;trophy=89;type=season")
soup = BeautifulSoup(response.content, 'html.parser')
urls = []
links = soup.find_all('a', class_='data-link', href=True)
for link in links:
    urls.append(link.get('href'))
filtered_urls = []

for url in urls:
  if "ci/engine/match" in url:
    filtered_urls.append(url)

print(len(filtered_urls))
filtered_urls
```

```python
res = requests.get("https://stats.espncricinfo.com/ci/engine/match/1298135.html")
soup = BeautifulSoup(res.content, 'html.parser')
div = soup.find_all('div', class_ = 'ds-flex ds-items-center ds-min-w-0 ds-mr-1')
div = div[-2].text +' vs '+ div[-1].text
span = soup.find_all('span', class_='ds-text-title-xs ds-font-bold ds-capitalize')
tables = soup.find_all('table', class_='ci-scorecard-table')
print(len(tables))
```

```python
tables[0]
```

```python
df = pd.read_html(str(tables[0]))
df = pd.DataFrame(df[0])
df['match'] = df.apply(lambda row: div, axis=1)
df['team'] = df.apply(lambda row: span[0].text, axis=1)
df.dropna(inplace=True,axis=0, how='all')
df
```

```python
data = pd.DataFrame(columns=['BATTING', 'Unnamed: 1', 'R', 'B', 'M', '4s', '6s', 'SR', 'Unnamed: 8',
       'Unnamed: 9', 'match', 'team'])
for url in filtered_urls:
    res = requests.get("https://stats.espncricinfo.com"+ url)
    soup = BeautifulSoup(res.content, 'html.parser')
    div = soup.find_all('div', class_ = 'ds-flex ds-items-center ds-min-w-0 ds-mr-1')
    div = div[-2].text +' vs '+ div[-1].text
    span = soup.find_all('span', class_='ds-text-title-xs ds-font-bold ds-capitalize')
    tables = soup.find_all('table', class_='ci-scorecard-table')
    c = 0
    for table in tables:
        df = pd.read_html(str(table))
        df = pd.DataFrame(df[0])
        df['match'] = df.apply(lambda row: div, axis=1)
        df['team'] = df.apply(lambda row: span[c].text, axis=1)
        c = c+1
        data = pd.concat([data,df], axis=0, ignore_index=True)
data       
```

```python
player_urls = []
for url in filtered_urls:
    res = requests.get("https://stats.espncricinfo.com"+ url)
    soup = BeautifulSoup(res.content, 'html.parser')
    links = soup.find_all('a', href=True)
    for link in links:
        if '/player/' in link.get('href'):
            player_urls.append(link.get('href'))
player_urls
```

```python
temp = pd.DataFrame(player_urls)
temp.drop_duplicates(inplace=True)
player_urls = temp.to_numpy().tolist()
player_urls
```

```python
from itertools import chain
player_urls = list(chain(*player_urls))
player_urls
```

```python
player_photos = []
for url in player_urls:
    res = requests.get("https://espncricinfo.com"+ url)
    soup = BeautifulSoup(res.content, 'html.parser')
    links = soup.find_all('a', href=True)
    for link in links:
        if '/photos' in link.get('href'):
            player_photos.append(link.get('href'))
temp = pd.DataFrame(player_photos)
temp.drop_duplicates(inplace=True)
player_photos = temp.to_numpy().tolist()
player_photos = list(chain(*player_photos))
player_photos
```

```python
df = pd.DataFrame()
response = requests.get('https://www.espncricinfo.com/player/finn-allen-959759')
soup = BeautifulSoup(response.content, 'html.parser')
name = soup.find_all('h1', class_='ds-text-title-l ds-font-bold')
team = soup.find_all('span', class_='ds-text-comfortable-s')
divs = soup.find_all('div', class_= 'ds-grid lg:ds-grid-cols-3 ds-grid-cols-2 ds-gap-4 ds-mb-8')
dict = {}
dict['name'] = name[0].text
dict['team'] = team[0].text
for div in divs:
    paras = div.find_all('p')
    headers = div.find_all('h5')
    for p, h in zip(paras, headers):
        dict[p.text] = h.text
    try:
        bat = dict['Batting Style']
    except:
        bat = None
    try:
        bowl = dict['Bowling Style']
    except:
        bowl = None
    try:
        role = dict['Playing Role']
    except:
        role = None
    temp = [[dict['name'], dict['team'], bat, bowl, role]]
    df1 = pd.DataFrame(temp, index=[0])
    df = pd.concat([df, df1], ignore_index=True, axis=0 )
df
```
